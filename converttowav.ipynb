```python
from pydub import AudioSegment
import os

def convert_to_wav(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith((".m4a", ".mp3", ".aac")):
            filepath = os.path.join(folder_path, filename)
            
            # Load the audio
            track = AudioSegment.from_file(filepath)
            
            # Export as .wav
            new_filename = os.path.splitext(filename)[0] + ".wav"
            track.export(os.path.join(folder_path, new_filename), format="wav")
            
            # Optional: Delete the old file to keep it clean
            os.remove(filepath)
            print(f"Converted: {filename} -> {new_filename}")


convert_to_wav('data/Am')
convert_to_wav('data/C')
convert_to_wav('data/E')
convert_to_wav('data/D')
convert_to_wav('data/A')
convert_to_wav('data/F')
convert_to_wav('data/Em')
convert_to_wav('data/G')

```
