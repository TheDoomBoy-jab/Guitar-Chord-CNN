```python
import os
import librosa
import soundfile as sf
import numpy as np

# --- SET YOUR PATHS HERE ---
original_data_dir = "data" # Where your current C, Dm, Em folders are
new_data_dir = "Augmented_Data" # Where the 200+ files will go

# The 8 chords you are using
label_names = ['Am', 'Bb', 'Bdim', 'C', 'Dm', 'Em', 'F', 'G']

# Create the new base directory if it doesn't exist
if not os.path.exists(new_data_dir):
    os.makedirs(new_data_dir)

def add_noise(data, noise_factor=0.005):
    noise = np.random.randn(len(data))
    return data + noise_factor * noise

def augment_and_save(file_path, chord_folder):
    # Load original audio
    y, sr = librosa.load(file_path, sr=48000)
    base_name = os.path.basename(file_path).replace('.wav', '')
    
    # 1. Save Original
    sf.write(f"{chord_folder}/{base_name}_orig.wav", y, sr)
    
    
    # 4. Time Stretch FAST (1.1x speed, fast strum)
    y_fast = librosa.effects.time_stretch(y, rate=1.1)
    sf.write(f"{chord_folder}/{base_name}_fast.wav", y_fast, sr)
    
    # 5. Time Stretch SLOW (0.9x speed, slow strum)
    y_slow = librosa.effects.time_stretch(y, rate=0.9)
    sf.write(f"{chord_folder}/{base_name}_slow.wav", y_slow, sr)
    
    # 6. Add Background Noise (Room hum)
    y_noise = add_noise(y)
    sf.write(f"{chord_folder}/{base_name}_noise.wav", y_noise, sr)

# --- RUN THE LOOP ---
for chord in label_names:
    print(f"Processing chord: {chord}...")
    
    # Make a subfolder for the chord in the new directory
    new_chord_folder = os.path.join(new_data_dir, chord)
    os.makedirs(new_chord_folder, exist_ok=True)
    
    # Get all .wav files in the original folder
    original_chord_folder = os.path.join(original_data_dir, chord)
    
    # Check if folder exists to avoid errors
    if os.path.exists(original_chord_folder):
        files = [f for f in os.listdir(original_chord_folder) if f.endswith('.wav')]
        
        for file in files:
            file_path = os.path.join(original_chord_folder, file)
            augment_and_save(file_path, new_chord_folder)
            
print("Augmentation Complete! Check the 'Augmented_Data' folder.")
```
